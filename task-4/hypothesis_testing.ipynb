# Hypothesis Testing: Critic Score vs Total Sales

## Objective
To analyze whether higher critic scores have a significant impact on total sales of video games.

## Hypothesis
Games with higher critic scores tend to achieve higher total sales.

## Method
Pearson correlation test is used to measure the relationship between critic scores and total sales.

## Expected Outcome
To determine whether a statistically significant relationship exists between the two variables.

```python
import pandas as pd
from scipy.stats import pearsonr

# Load dataset
df = pd.read_csv("Video_Games_Sales_Cleaned.csv")

# Remove missing values
df = df.dropna(subset=['critic_score', 'total_sales'])

# Apply correlation test
corr, p_value = pearsonr(df['critic_score'], df['total_sales'])

print("Correlation:", corr)
print("P-value:", p_value)
```

## Visualization: Critic Score vs Total Sales

This scatter plot illustrates the relationship between critic scores and total sales of video games. It helps visually assess whether higher critic ratings are associated with increased sales.

```python
import matplotlib.pyplot as plt

plt.scatter(df['critic_score'], df['total_sales'])
plt.xlabel("Critic Score")
plt.ylabel("Total Sales")
plt.title("Critic Score vs Sales")
plt.show()
```
